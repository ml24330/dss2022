```python
import pandas as pd
from sklearn.model_selection import KFold 
from sklearn.metrics import confusion_matrix
from sklearn import svm

uk_clocks = pd.read_csv('uk_clocks.csv')
uk_clocks = uk_clocks[uk_clocks.isSold == 1].reset_index(drop=True)
```

```python
predictors = ['listingIsTopRated', 'sellerFeedbackScore', 'sellerPositivePercent', 'sellerIsTopRated', 'endAtWeekend', 'endAtEvening', 'length', 'isBroken', 'isUsed', 'isLarge', 'freeShipping']

X = uk_clocks.loc[:,predictors]
y = uk_clocks.loc[:,'price']
```

```python
X
```

```python
from numpy import mean
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.model_selection import RepeatedKFold
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from sklearn.metrics import balanced_accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.preprocessing import StandardScaler


param_grid = {
    "model__n_estimators": [5, 30, 80, 150, 300],
    "model__max_depth": [1, 3, 7, 14, 20, 30]
}

model = GradientBoostingRegressor()
steps = [("scaler", StandardScaler()), ('model', model)]
pipeline = Pipeline(steps=steps)
cv = RepeatedKFold(n_splits=5, n_repeats=3, random_state=1)
search = GridSearchCV(pipeline, param_grid, cv=cv, n_jobs=-1)
search.fit(X, y)
print("Best parameter (CV score=%0.3f):" % search.best_score_)
print(search.best_params_)
```

```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

model = GradientBoostingRegressor(max_depth=7, n_estimators=5)
steps = [("scaler", StandardScaler()), ('model', model)]
pipeline = Pipeline(steps=steps)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

pipeline.fit(X_train,y_train)
y_pred = pipeline.predict(X_test)
plt.xlim(0,200)
plt.ylim(0,200)
plt.scatter(y_test, y_pred)
```

```python
import numpy as np

# parameters
U = lambda x: np.log(x+1)
D = 0.82
B = 0.66

res0 = {}
res1 = {}

# iterate through all possible consumption bundles
for c1 in range(11):
    for c2 in range(11-c1):
        c3 = 10-c1-c2
        U0 = U(c1)+D*U(c2)+D**2*U(c3)
        U1 = U(c1)+B*D*U(c2)+B*D**2*U(c3)
        res0[f'{c1},{c2}'] = U0
        res1[f'{c1},{c2}'] = U1
        
# calculate maximum utility and corresponding bundles
m0 = np.array(list(res0.values())).max()
m1 = np.array(list(res1.values())).max()
k0 = list(res0.keys())[list(res0.values()).index(m0)]
k1 = list(res1.keys())[list(res1.values()).index(m1)]

# display results
print(f'''At day 0, Joel chooses to consume {k0[0]} eggs in day 1, {k0[-1]} eggs in day 2, {10-int(k0[0])-int(k0[-1])} 
      eggs in day 3 to get utility {m0:.5f}.''')
print(f'''At day 1, Joel chooses to consume {k1[0]} eggs in day 1, {k1[-1]} eggs in day 2, {10-int(k1[0])-int(k1[-1])} 
eggs in day 3 to get utility {m1:.5f}.''')
```
