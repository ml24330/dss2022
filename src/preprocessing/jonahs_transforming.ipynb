```python
import pandas as pd
from db_utils import Query, QueryAll, QueryAllFinished, UpdateItem

data = QueryAll("clock_uk")
cols = ["id", "title", "globalId", "categoryId", "categoryName", "url", "location", "shippingType", "shippingLocations",
                "shippingTime", "startTime", "endTime", "returnsAccepted", "conditionId", "listingIsTopRated", 
                "sellerFeedbackScore", "sellerPositivePercent", "sellerName", "sellerIsTopRated", "price", "currency", "bids"]
```

```python
df = pd.DataFrame(data, columns=cols)
df
```

```python
##Create a new df with all the qualitative data
new_df = pd.DataFrame()
new_df['title'] = df['title']
new_df['free_shipping'] = (df['shippingType'] == 'Free').astype(int)


end_of_week = ['2022-10-01', '2022-10-02', '2022-10-07', '2022-10-08', '2022-10-09', '2022-10-14', '2022-10-15', '2022-10-16', '2022-10-21', '2022-10-22', '2022-10-23', '2022-10-28', '2022-10-29', '2022-10-30', '2022-11-04', '2022-11-05', '2022-11-06', '2022-11-11', '2022-11-12', '2022-11-13', '2022-11-18', '2022-11-19', '2022-11-20','2022-11-25','2022-11-26','2022-11-27']
used = [3000, 6000, 2500, 5000, 2750]
new = ['1000', '1500']
broken = [7000]

new_df['starts_at_week_end'] = df["startTime"].str.slice(0, 10).isin(end_of_week).astype(int)
new_df['end_at_weeks_end'] = df["endTime"].str.slice(0, 10).isin(end_of_week).astype(int)

new_df['start_part_of_day'] = df["startTime"].str.slice(11, 13).astype(int).gt(11).astype(int)+df["startTime"].str.slice(11, 13).astype(int).gt(18).astype(int)
new_df['end_part_of_day'] = df["endTime"].str.slice(11, 13).astype(int).gt(11).astype(int)+df["endTime"].str.slice(11, 13).astype(int).gt(18).astype(int)

new_df['broken'] = df['conditionId'].isin(broken).astype(int)
new_df['used'] = df['conditionId'].isin(used).astype(int)
new_df['listingTopRated'] = df['listingIsTopRated']

new_df["sellerFeedbackScore"] = df['sellerFeedbackScore']

new_df['sellerPositivePercent'] = df['sellerPositivePercent']

new_df['sellerIsTopRated'] = df['sellerIsTopRated']

df['price'] = df['price'].str.replace('.',',', regex=False)

df['price'].str.slice(1,).str.isalnum()#.str.isnumeric()
```
