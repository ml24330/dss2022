```python
import matplotlib.pyplot as plt
import pandas as pd
import re
import numpy as np
from scipy import stats
from db_utils import Query, QueryAll, QueryAllFinished, UpdateItem

data = QueryAll("clock_uk")
cols = ["id", "title", "globalId", "categoryId", "categoryName", "url", "location", "shippingType", "shippingLocations",
                "shippingTime", "startTime", "endTime", "returnsAccepted", "conditionId", "listingIsTopRated", 
                "sellerFeedbackScore", "sellerPositivePercent", "sellerName", "sellerIsTopRated", "price", "currency", "bids"]
```

```python
clocks = pd.DataFrame(data, columns = cols)
clocks
```

```python
#remove missing values from bids
clocks = clocks[clocks.bids >= 0]

#transform price column into numerical column
def find_number(text):
    num = re.findall("\d+\.\d+",text)
    return " ".join(num)
clocks['price']=pd.to_numeric(clocks['price'].apply(lambda x: find_number(x)))

#remove outliers from numerical column
clocks = clocks[(np.abs(stats.zscore(clocks['sellerFeedbackScore'])) < 3)]
clocks = clocks[(np.abs(stats.zscore(clocks['sellerPositivePercent'])) < 3)]
clocks = clocks[(np.abs(stats.zscore(clocks['price'])) < 3)]
clocks = clocks[(np.abs(stats.zscore(clocks['bids'])) < 3)]

clocks
```
